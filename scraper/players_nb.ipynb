```python
import requests
import re
import time
```

```python
url = "https://liquipedia.net/valorant/"
teamUrl = []
```

```python
def getSrc(url,id):
    r = requests.get(url+id)
    return r.text
```

### Team Name(Url) Scraping

```python
def getTeamUrl():
    active = []
    portal = 'Portal:Teams'
    src = getSrc(url, 'Portal:Teams')
    
    finder = '<tr>.+?<span class="team-template-text"><a href="/valorant/(.+?)" title'

    for match in re.findall(finder,src):
        if match not in active:
            active.append(match)
    return active
```

### Player Name(Url) Scraping

```python

portal = "Portal:Players/"
regions = ["Europe","CIS","America","Oceania","Asia","Africa_%26_Middle_East"]
# regions = ["Europe"]
```

```python
def getPlayerUrl():
    active = []
    for region in regions:

        src = getSrc(url,portal+region)

        tablefinder = '<table .+?<tbody><tr><th.+?</a></span> (Players</th></tr><tr>.+?</td></tr>)</tbody></table>'

        for match in re.findall(tablefinder,src):

            finder = '<tr><td style="text-align:left"> <span class="flag">.+?</span>.+?<a href="/valorant(.+?)" title='
            for playermatch in re.findall(finder,match):
                if playermatch not in active:
                    active.append(playermatch)
        time.sleep(3)

    return active
    
```

### Player Initial Information Scraping

```python
src = None
portal = 'Portal:Teams'
```

```python
def getPlayerInfo():
    players = []
    global src
    if not src:
        src = getSrc(url,portal)
    teamFinder = '<table .+?><tbody>(.+?)</tbody></table>'
    
    for teamMatch in re.findall(teamFinder,src):
        # print(teamMatch)
        teamNameFinder = '<tr>.+?<span class="team-template-text"><a href="/valorant/(.+?)" title'
        teamName = re.findall(teamNameFinder,teamMatch)

        playerFinder = '<tr><td style="font-weight:bold"><span class="flag"><a .+?(<a href="/valorant/.+?".*?title=".+?">.+?</a></span></td><td>.*?</td></tr>)'
        for playerMatch in re.findall(playerFinder,teamMatch):
            # print(playerMatch)
            player = {
                "url": None,
                "name": None,
                "realName": None,
                "team": None
            }
            playerUrl = re.findall('<a href="/valorant/(.+?)" title=',playerMatch)[0]
            player["url"] = re.findall('<a href="/valorant/(.+?)" title=',playerMatch)[0] if playerUrl[0:5] != "index" else None
            player["name"] = re.findall('title=".+?">(.+?)</a></span>',playerMatch)[0]
            playerRName = re.findall('</a></span></td><td>(.*?)</td></tr>',playerMatch)[0]
            player["realName"] = playerRName if playerRName != "" else None
            player["team"] = teamName[0] if len(teamName) > 0 else 'Persija Esports'
            players.append(player)
            
    return players

```

```python
players = getPlayerInfo()
```

## Player Additional Data Scraping

```python
from datetime import datetime
import time
import random
import urllib.request
```

```python
# Sleep random time between [a, b]
def sleep_interval(a, b):
    time.sleep(random.randint(a, b))
```

```python
addedPlayerData = {
    "imageUrl": None,
    "country": [],
    "born": None,
    "mainAgent": [],
}
gamingGear = {
    "mouse": [],
    "keyboard": [],
    "headset": [],
    "monitor": [],
    "mousepad": [],
}
addedPlayerData.update({"gamingGear":gamingGear})
addedPlayerData
```

```python
src = None
PLAYERPATTERN = {
    "PlayerCard":'<div class="fo-nttax-infobox-wrapper infobox-valorant">(.+?)<div class="fo-nttax-infobox-adbox wiki-bordercolor-light">',
    "imageUrl":'<div class="infobox-image">.+?<a href="/valorant/.+?" class="image"><img alt="" src="(.+?)"',
    "born": '<span style="display:none"> \(<span class="bday">(.+?)</span>\) </span>',
    "country": '<span class="flag"><a .+?</a></span> <a .+?>(.+?)</a>',
    "agentPanel": 'Main Agents:</div>\n<div class="infobox-cell-2">.+?</div>',
    "agent": '<a href="/valorant/(.+?)" title=".+?">',
    "hardwarePanel":'<th>Hardware.+?</div>',
    "hardware": '<b>(?!</b>|<small>)(?:<a.+?>|)(.+?)(?:</a>|)</b>',
    "brand":'<img alt="" src="/commons/images/.+?/.+?/(.+?)(?:-icon.png"|logo_std)',
    "hardwaretype": '>(Mouse(?:pad)?|Monitor|Keyboard|Headset)'
}
```

```python
noImageString = "/commons/images/1/19/Infobox_player_NoImage.png"
```

```python
def getImageFile(imageUrl,path):
    urllib.request.urlretrieve("https://liquipedia.net"+imageUrl, "data/images/players/"+path+"."+imageUrl.rsplit('.')[-1])
```

```python
def getMorePlayerInfo(player):
    addPlayerData = {
        "imageUrl": None,
        "country": [],
        "born": None,
        "mainAgent": [],
    }
    gamingGear = {
        "mouse": None,
        "mousepad": None,
        "monitor": None,
        "keyboard": None,
        "headset": None,
    }

    if player["url"]:
        src = getSrc(url, player["url"])
        playerCardMatch = re.findall(
            PLAYERPATTERN["PlayerCard"], src, re.DOTALL)
        if playerCardMatch[0]:
            playerImageUrl = re.findall(
                PLAYERPATTERN["imageUrl"], playerCardMatch[0], re.DOTALL)
            if len(playerImageUrl) > 0:
                if playerImageUrl[0] != noImageString:
                    addPlayerData["imageUrl"] = playerImageUrl[0]
                    sleep_interval(1, 3)
                    getImageFile(playerImageUrl[0], player["url"])

            playerBorn = re.findall(
                PLAYERPATTERN["born"], playerCardMatch[0], re.DOTALL)
            addPlayerData["born"] = playerBorn[0] if len(playerBorn) > 0 else None

            for playerCountry in re.findall(PLAYERPATTERN["country"], playerCardMatch[0], re.DOTALL):
                addPlayerData["country"].append(playerCountry)
                
            # print(playerCardMatch[0])
            agentPanel = re.findall(
                PLAYERPATTERN["agentPanel"], playerCardMatch[0], re.DOTALL)
            if len(agentPanel) > 0 :
                for playerAgent in re.findall(PLAYERPATTERN["agent"], agentPanel[0], re.DOTALL):
                    # print(playerAgent)
                    addPlayerData["mainAgent"].append(playerAgent)

            
            hardwarePanel = re.findall(PLAYERPATTERN["hardwarePanel"], src, re.DOTALL)
            if len(hardwarePanel)>0:
                hardwareList = []
                brandList =[]
                hardwareTypeList = []
                # print(hardwarePanel)
                for hardwareType in re.findall(PLAYERPATTERN["hardwaretype"], hardwarePanel[0]):
                    hardwareTypeList.append(hardwareType)

                for hardware in re.findall(PLAYERPATTERN["hardware"],hardwarePanel[0]):
                    hardwareList.append(hardware)
                
                for brand in re.findall(PLAYERPATTERN["brand"],hardwarePanel[0]):
                    brandList.append(''.join(brand))

                # print(hardwareTypeList,"\n",hardwareList,"\n",brandList)
                if len(hardwareTypeList) != len(hardwareList) or len(hardwareTypeList) != len(brandList):
                    for index,item in enumerate(brandList):
                        if item == "Unknown":
                            hardwareTypeList.pop(index)
                            brandList.pop(index)

                for index in range(len(hardwareTypeList)):
                    gamingGear[hardwareTypeList[index].lower()] = {
                        "brand": brandList[index],
                        "name": hardwareList[index]
                    }
        else:
            print("No player card found")
            return 
            
        
        addPlayerData.update({"gamingGear":gamingGear})
        sleep_interval(4, 8)
        # print(addPlayerData)
    player.update(addPlayerData)
```

```python
#main scrapping process
progress = 735
playercount = len(players)
for index,player in enumerate(players[progress:playercount]):
    getMorePlayerInfo(player)
    print(player["name"],"Information is updated","("+str(progress+index+1)+"/"+str(len(players[:playercount]))+")")
      
```
